```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.datasets import mnist
```

### Get the data and Preprocess it

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
def plot_input_image(i):
    plt.imshow(X_train[i], cmap='binary')
    plt.title(y_train[i])
    plt.show()
```

```python
for i in range(10):
    plot_input_image(i)
```

```python
# Pre processing the images

# Normalizing the image to [0, 1] range
X_train = X_train.astype(np.float32)/255
X_test = X_test.astype(np.float32)/255
```

X_train = X_train.astype(np.float32)/255
X_test = X_test.astype(np.float32)/255
In this code snippet, X_train and X_test are NumPy arrays that represent the input data for training and testing the machine learning model, respectively. The code performs the following operations on these arrays:

X_train.astype(np.float32): This converts the data type of the elements in the X_train array to np.float32. np refers to the NumPy library that provides mathematical functions and tools for working with arrays.
/255: This divides every element in the X_train array by 255. This step is a form of normalization, where the pixel values are scaled down to a range between 0 and 1. In the MNIST dataset, the pixel values range from 0 to 255, with 0 being black and 255 being white.
The same operations are then performed on the X_test array to ensure consistent preprocessing for the test data.

By dividing the pixel values by 255, the code effectively scales down the pixel intensity values to a normalized range, which can be beneficial for training machine learning models.

In this context, the term "data type" refers to how the values in the array are represented in computer memory. Different data types have different ranges and precision levels. For example, integers (e.g., int, int32) represent whole numbers without decimal places, while floating-point numbers (e.g., float, float32) can represent decimal values as well.

The X_train.astype(np.float32) part of the code uses the astype() method in NumPy to convert the data type of the elements in the X_train array to np.float32. The np.float32 is a NumPy data type representing single-precision floating-point numbers.

Converting the data type to np.float32 is a common practice in machine learning tasks because many machine learning algorithms work best with floating-point data. It ensures that the data is in a format suitable for numerical computations and helps maintain consistency in the data representation during training and inference.

```python
# Reshape / expand the dimentions of images to (28,28,1)
X_train = np.expand_dims(X_train, -1)
X_test = np.expand_dims(X_test, -1)
```

In this code, X_train and X_test are NumPy arrays that represent the input data for training and testing the ML model, respectively. The np.expand_dims() function is used to modify these arrays.

The np.expand_dims() function is employed to add an extra dimension to the input data. In this case, -1 is passed as the second argument to indicate where this extra dimension should be added. The -1 argument in np.expand_dims() ensures that the new dimension is added as the last dimension of the array.

Adding an extra dimension can be useful in scenarios where the input data needs to be compatible with certain ML models or frameworks that expect a particular shape or dimensionality. For example, convolutional neural networks (CNNs) typically require input data in the shape of (batch_size, height, width, channels), where channels represents the number of color channels (e.g., 1 for grayscale or 3 for RGB). By using np.expand_dims() with -1, the code is adding an extra dimension to the arrays, making them compatible with models that expect a single color channel.

By expanding the dimensions, the code ensures that the input data has the required shape for further processing or model training.

```python
# Convert classes to one hot vector
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
```

```python
y_test
```

The code snippet you provided is related to the preprocessing step for the target labels in an ML project that aims to identify handwritten numbers using the MNIST dataset. The code utilizes the to_categorical() function from the Keras library to convert the target labels into one-hot encoded vectors. Here's a breakdown of the code:

y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
In this code, y_train and y_test represent the target labels or classes for the training and testing data, respectively. The to_categorical() function from Keras is used to convert these labels into one-hot encoded vectors.

One-hot encoding is a common technique used in classification tasks, where each class is represented as a binary vector. The length of the vector is equal to the number of classes, and the vector has a value of 1 at the index corresponding to the class label, while all other positions are filled with 0s.

By using to_categorical(), the code converts the original labels, which are integers representing the class labels (e.g., 0, 1, 2, ..., 9 for the MNIST dataset), into one-hot encoded vectors. This is necessary because many ML algorithms and neural network models expect the target labels to be in a one-hot encoded format.

The resulting y_train and y_test arrays will have the shape (number of samples, number of classes) and contain the one-hot encoded vectors representing the class labels for each sample in the training and testing data, respectively. This format makes it easier to train ML models and evaluate their performance for classification tasks.

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout
```

```python
# Build the model
model = Sequential()

model.add(Conv2D(32, (3,3), input_shape = (28,28,1), activation = 'relu'))
model.add(MaxPool2D((2,2)))

model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(MaxPool2D((2,2)))

model.add(Flatten())

model.add(Dropout(0.25))

model.add(Dense(10, activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = keras.losses.categorical_crossentropy, metrics = ['accuracy'])
```

```python
# Callbacks
from keras.callbacks import EarlyStopping, ModelCheckpoint

# Earlystopping
es = EarlyStopping(monitor='val_acc', min_delta=0.01, patience=4, verbose=1)

# Model Check Point
mc = ModelCheckpoint("./bestmodel.h5", monitor="val_acc", verbose=1, save_best_only=True)

cb = [es, mc]
```

from keras.callbacks import EarlyStopping, ModelCheckpoint: This line imports the EarlyStopping and ModelCheckpoint callback classes from the Keras library. Callbacks are objects used to customize the behavior of the training process in Keras.

es = EarlyStopping(monitor='val_acc', min_delta=0.01, patience=4, verbose=1): This line initializes an EarlyStopping callback object. The EarlyStopping callback is used to stop the training process early if a certain criterion is met. The arguments provided are:

monitor='val_acc': The quantity to monitor during training. In this case, it's the validation accuracy (val_acc).
min_delta=0.01: The minimum change in the monitored quantity that qualifies as an improvement.
patience=4: The number of epochs to wait before stopping the training if no improvement is observed.
verbose=1: The verbosity level. It determines the amount of information displayed during training.
mc = ModelCheckpoint("./bestmodel.h5", monitor="val_acc", verbose=1, save_best_only=True): This line initializes a ModelCheckpoint callback object. The ModelCheckpoint callback saves the best model during training based on a certain criterion. The arguments provided are:

"./bestmodel.h5": The filepath where the best model will be saved.
monitor="val_acc": The quantity to monitor during training to determine the best model. In this case, it's the validation accuracy (val_acc).
verbose=1: The verbosity level. It determines the amount of information displayed during training.
save_best_only=True: Specifies whether to save only the best model based on the monitored quantity. If set to True, only the model with the best validation accuracy will be saved.
These callbacks can be used during the training process to monitor the model's performance, early stop if necessary, and save the best model based on a specified metric.

### Model Training

```python
his = model.fit(X_train, y_train, epochs=15, validation_split=0.5, callbacks=cb)
```

In this code, model refers to the ML model that has been defined and compiled prior to this training step. X_train and y_train represent the input features and target labels, respectively, for the training data.

The fit() method is a fundamental method in Keras that performs the training process for a model. It trains the model on the provided data for a specified number of epochs. Here are the arguments used in the fit() method:

X_train: The input features (input data) for training the model.
y_train: The target labels (output data) corresponding to the input features.
epochs=5: The number of times the entire training dataset will be iterated over during training. In this case, the model will be trained for 5 epochs.
validation_split=0.3: The portion of the training data to be used for validation during training. In this case, 30% of the training data will be used for validation, and the remaining 70% will be used for actual training.
callbacks=cb: This argument is used to provide callback objects that customize the training process. The cb variable is assumed to contain the callbacks, such as EarlyStopping and ModelCheckpoint, that were defined earlier.
The fit() method returns a History object (his in this case) that contains information about the training process, such as the loss and accuracy values at each epoch. This object can be used to analyze and visualize the training progress.
